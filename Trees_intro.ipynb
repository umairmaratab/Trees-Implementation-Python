```python
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
```

```python
# Traversals
# T L R   Pre-order Traversal
# L R T   Post-order Traversal
# L T R   in-order   Traversal
```

```python
# This dfs is implemented on Pre-order Traversal
def dfs_pre(self):
    print(self.val)
    
    if self.left:
        self.left.dfs_pre()
    if self.right:
        self.right.dfs_pre()
        
TreeNode.dfs_pre = dfs_pre
```

```python
def dfs_post(self):
    if self.left:
        self.left.dfs_post()
    if self.right:
        self.right.dfs_post()
    print(self.val)
    
TreeNode.dfs_post = dfs_post
```

```python
def dfs_inorder(self):
    if self.left:
        self.left.dfs_inorder()
    print(self.val)
    
    if self.right:
        self.right.dfd_inorder()
    
TreeNode.dfs_inorder = dfs_inorder
```

```python
def bfs(self):
    
    to_visit = [self]
    
    while to_visit:
        current = to_visit.pop(0)
        
        print(current.val)
        
        if current.left:
            to_visit.append(current.left)
        if current.right:
            to_visit.append(current.right)
            
TreeNode.bfs = bfs
```

```python
def dfs_apply(self, fn):
    fn(self)
    
    if self.left:
        self.left.dfs_apply(fn)
    if self.right:
        self.right.dfs_apply(fn)
        
TreeNode.dfs_apply = dfs_apply
```

```python
class PerformSum:
    def __init__(self):
        self.sum = 0
        
    def process(self, node):
        self.sum += node.val
        
    def get_sum(self):
        return self.sum
    
    def reset_sum(self):
        self.sum = 0
        
p = PerformSum()
```

```python
p.reset_sum()
t1.dfs_apply(p.process)
print(p.get_sum())
```

```python
from graphviz import Digraph

def visualize_tree(tree):
    if tree is None: return 'Nothing in the tree!'
    def add_nodes_edges(tree, dot=None):
    # Create Digraph object
    
        if dot is None:
            dot = Digraph()
            dot.attr('node', shape='circle')
            dot.node(name=str(tree), label=str(tree.val))
        for child in [tree.left, tree.right]: # do for all children
            if child is not None:
                if child == tree.left: dot.attr('node', shape='circle', style='filled',
                fillcolor='lightblue')
                if child == tree.right: dot.attr('node', shape='doublecircle', style='filled',
                fillcolor='seashell')
                dot.node(name=str(child) ,label=str(child.val))
                dot.edge(str(tree), str(child))
                dot = add_nodes_edges(child, dot=dot) # recursive call
        return dot
    
    # Add nodes recursively and create a list of edges
    dot = add_nodes_edges(tree)

    # Visualize the graph
    display(dot)
def print_tree(tree, level=0, label='.'):
    print(' ' * (level*2) + label + ':', tree.val)
    for child, lbl in zip([tree.left, tree.right], ['L', 'R']):
        if child is not None:
            print_tree(child, level+1, lbl)
    
```
